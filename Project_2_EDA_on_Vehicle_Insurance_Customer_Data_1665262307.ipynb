A company has customer data that contains 8 columns of customer details and another table having name customer_policy data contains the policy details of the customer. 

The company intends to offer some discount in premium for certain customers. To do that they ask their Data scientist team to get some information. Hence, following tasks DS team decided to perform:

## 1. Add the column names to both datasets:

```python
import pandas as pd
import numpy as np
```

```python
customer=pd.read_csv("customer_details.csv")
customer.columns=[ "customer_id","Gender","age","driving licence present","region code","previously insured","vehicle age","vehicle damage"]

customer_policy=pd.read_csv("customer_policy_details.csv")
customer_policy.columns=["customer_id","annual premium (in Rs)","sales channel code","vintage","response"]
customer_policy
```

## 2. Checking and Cleaning Data Quality:


### i. Null values

Generate a summary of count of all the null values column wise Drop Null values for customer_id because central tendencies for id’s is not feasible. Replace all null values for numeric columns by mean. Replace all null values for Categorical value by mode.

```python
# Null values
print("Summary of count of all the null values column wise for the table customer")
print(customer.isna().sum())
print("\n")
print("Summary of count of all the null values column wise for table customer_policy")
print(customer_policy.isna().sum())
```

```python
# Drop null custumer_id
customer.dropna(subset=["customer_id"],inplace=True)
customer_policy.dropna(subset=["customer_id"],inplace=True)

#filna with mean()
customer.fillna(customer.mean(axis=0,numeric_only=True),inplace=True)
customer_policy.fillna(customer_policy.mean(axis=0,numeric_only=True),inplace=True)

#fillna with mode()
customer.fillna(customer.mode(axis=0,numeric_only=False).iloc[0,:],inplace=True)
customer_policy.fillna(customer_policy.mode(axis=0,numeric_only=False).iloc[0,:],inplace=True)
```

### ii. Outliers

Generate a summary of count of all the outliers column wise Replace all outlier values for numeric columns by mean.

(Hint1: for outlier treatment use IQR method as follows:

For example: for a column X calculate Q1 = 25th percentile and Q3 = 75th percentile then IQR = Q3 – Q1 ) then to check outlier, anything lower than a Q1 – 1.5IQR or greater than Q3 + 1.5 IQR would be an outlier

Hint2: For getting percentile value, explore pd.describe() function)

```python
def find_outliers_IQR(df):
   q1=df.quantile(0.25)
   q3=df.quantile(0.75)
   IQR=q3-q1
   print("IQR",IQR)
   inliers = df[~((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
   print("Valid data shape",outliers.shape)
   return inliers

print("##### Customer table ####")
outlier_customer = find_outliers_IQR(customer)
print("Number of outliers in customer table : "+ str(len(outlier_customer)))
print("max outlier value in customer table: "+ str(outlier_customer.max()))
print("min outlier value in customer table: "+ str(outlier_customer.min()))
print("\n")

print("##### Customer_policy table ####")
outlier_customer_policy = find_outliers_IQR(customer_policy)
print("Number of outliers in customer table : "+ str(len(outlier_customer_policy)))
print("max outlier value in customer table: "+ str(outlier_customer_policy.max()))
print("min outlier value in customer table: "+ str(outlier_customer_policy.min()))
print("\n")
```

```python
# Putting mean value in outliers value
outlier_customer_policy.fillna(outlier_customer_policy.mean(numeric_only=True,axis=0),inplace=True)
outlier_customer.fillna(outlier_customer.mean(numeric_only=True,axis=0),inplace=True)
customer=outlier_customer
customer_policy=outlier_customer_policy
customer
```

### iii. White spaces Remove white spaces

```python
# Remove whitespaces
for i in range(customer.shape[1]-1):
  if customer.iloc[:,i].dtype =="str":
    customer.iloc[:,i]=customer.iloc[:,i].str.strip()

for i in range(customer_policy.shape[1]-1):
  if customer_policy.iloc[:,i].dtype =="str":
    customer_policy.iloc[:,i]=customer_policy.iloc[:,i].str.strip()
```

### iv. case correction(lower or upper, any one)

```python
customer=customer.apply(lambda x:x.str.lower() if x.dtypes=="str" else x)
customer_policy=customer_policy.apply(lambda x:x.str.lower() if x.dtypes=="str" else x)
```

### v. Convert nominal data (categorical) into dummies for future modeling use if required

```python
customer_dummy=pd.get_dummies(customer)
customer_dummy
customer_policy_dummy=pd.get_dummies(customer_policy)
customer_policy_dummy
```

### vi. Drop Duplicates (duplicated rows)

```python
customer.drop_duplicates(inplace=True)
customer_policy.drop_duplicates(inplace=True)
```

## 3 Create a Master table for future use. Join the customer table and customer_policy table to get a master table using customer_id in both tables.

(Hint: use pd.merge() function)

```python
Master=pd.merge(customer,customer_policy,on="customer_id")
Master
```

## 5. Is there any relation between Person Age and annual premium?
Hint: use correlation function (Correlation describes the relationship between two variables). 

Correlation coefficient < -0.5           - Strong negative relationship

Correlation coefficient > 0.5            -  Strong positive relationship

0.5 < Correlation coefficient < 0.5   - There is no relationship. 

```python
correaltion_coefficient=np.corrcoef(Master["age"],Master["annual premium (in Rs)"])
print("Correlation coefficient between Person Age and annual premium is {:.4f}".format(correaltion_coefficient[1][0]))
if correaltion_coefficient[1][0]>0.5:
  print("Strong positive relationship")
elif correaltion_coefficient[1][0]<-0.5:
  print("Strong negative relationship")
else:
  print("There is no relationship")
```
